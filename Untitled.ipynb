# 2.d. 每個月參Tweets數量長條圖

```python
q = pyes.query.MatchQuery("text","security OR attack OR CVE OR Vulnerability OR Exploit OR hacker OR botnet OR malware OR APT OR hat")
tagg = pyes.aggs.DateHistogramAgg('tweets_over_time', field= 'created_at',interval="1M",format="EEE MMM dd HH:mm:ss ZZZZZ yyyy")

qsearch = pyes.query.Search(q)

qsearch.agg.add(tagg)

result_tweets_by_month = conn.search(query=qsearch , doc_types='tweet') 
```

```python
from datetime import datetime
import calendar
```

```python
tweet_list = []
for bucket in result_tweets_by_month_list:
    temp = {}
    dt = datetime.strptime(bucket['key_as_string'], '%a %b %d %H:%M:%S +0000 %Y')
    temp['month']= calendar.month_abbr[dt.month]+','+str(dt.year)
    temp['tweet_count'] = bucket['doc_count']
    tweet_list.append(temp)
```

```python
bq = pyes.query.BoolQuery() 

ESR = pyes.ESRange(field="created_at", from_value="Thu Jan 01 00:00:00 +0000 2009", to_value="Thu Dec 31 23:59:59 +0000 2009")
rq = pyes.query.RangeQuery(qrange=ESR)
tq = pyes.query.TermQuery(field="uid", value="21501463") 
bq.add_must(rq)
bq.add_must(tq)
result_2009 = conn.search(query=bq , doc_types='tweet') 
```

```python
def combined_text(year,uid):
    fromvalue=''
    tovalue=''
    if year==2009:
        fromvalue='Thu Jan 01 00:00:00 +0000 2009'
        tovalue="Thu Dec 31 23:59:59 +0000 2009"
    elif year==2012:
        fromvalue='Sun Jan 01 00:00:00 +0000 2012'
        tovalue="Mon Dec 31 23:59:59 +0000 2012"
    elif year==2015:
        fromvalue='Thu Jan 01 00:00:00 +0000 2015'
        tovalue="Thu Dec 31 23:59:59 +0000 2015"
    
    bq = pyes.query.BoolQuery() 
    ESR = pyes.ESRange(field="created_at", from_value=fromvalue, to_value=tovalue)
    rq = pyes.query.RangeQuery(qrange=ESR)
    tq = pyes.query.TermQuery(field="uid", value=uid) 
    bq.add_must(rq)
    bq.add_must(tq)
    result = conn.search(query=bq , doc_types='tweet')
    combined_text=''
    for i in result:
        combined_text=combined_text+i['text']
    return combined_text
```

```python
combined_text(2015,"14669471")
```

```python
def year_text(year,observed_uid):
    year_text=''
    for i in observed_uid:
        if i['year']==year:
            for uid in i['uid']:
                year_text=year_text+combined_text(year,uid)
    return year_text
```

```python
def uid_keyword(year,uid):
    uid_txt = combined_text(year,uid)
    Rake = rake.Rake("/Users/ning_kang/Desktop/school/social_media_analysis/SmartStoplist.txt",3,3,2);
    keywords_uid = Rake.run(uid_txt)
    return keywords_uid
```

```python
uid_keyword(2009,"21501463")
```

```python
def twitter_data(year,uid,feature_names):
    tmp={}
    tmp['uid']=uid
    tmp['label']=(year-2009)/3
    tmp['data']=[]
    uid_keyword_list = uid_keyword(year,str(uid))
    i=0
    for f in feature_names:
        tmp['data'].append(0)
        for k in uid_keyword_list:
            if f == k[0]:
                tmp['data'][i]=k[1]
        i=i+1
    return tmp
```

```python
twitter_data(2009,21501463,twitter_feature_names)
```

```python
data_array=[]
target_array=[]

for bucket in observed_uid:
    year = bucket['year']
    for uid in bucket['uid']:
        dic=twitter_data(year,uid,twitter_feature_names)
        data_array.append(dic['data'])
        target_array.append(dic['label'])
```
